```python
import pandas as pd
import json
from serpapi import google_scholar_search as gs
import os
import time
from dotenv import load_dotenv

load_dotenv()
```

```python
alternatives = {
'reliance' : [
    'reliance',
    '"appropriate reliance"',
    '"AI reliance"',
    'under-reliance',
    'over-reliance'
],

'decision_making' : [
    '"decision making"',
    '"decision-making"',
],

'human' : [
    'human',
    'user',
    'participant',
    'participants',
    'users',
    'humans',
    '"lay person"',
    '"lay user"'
],

'ai' : [
    'artificial intelligence',
    'ai',
    'AI',
    'ML',
    'LLM',
    'NN',
    '"neural networks"',
    '"machine learning"',
    '"Large Language Model"',
    '"deep learning"'
],

'collaboration' : [
    'assisted',
    'assist',
    'collaboration',
    'help'
],

'experiment' : [
    'empirical',
    '"user study"',
    'experiment',
    '"crowdsourced study"',
],

# 'cognition' : [
#     'attitude',
#     'belief',
#     'beliefs',
# ],

# 'visualization' : [
#     'visualization',
# ]
}
```

```python
def create_search_query(keywords, source):
    temps = []
    for item in keywords:
        temps.append(f"({' OR '.join(keywords[item])})")
    res = (" ".join(temps))
    print(res)
    return res
```

```python
search_query = create_search_query(alternatives, 'acm')
```

```python

api_key = os.getenv('SERPAPI_KEY')
```

```python
num_pages = 70
all_results = []


params = {
    "engine": "google_scholar",
    "q": f"{search_query} site:acm.org",  # Replace with your query
    "as_ylo": "2020",  # Minimum year
    # "as_yhi": "2024",  # Maximum year
    "source": "ACM",  # Specific source if available (e.g., "Nature")
    "api_key": api_key,
    'hl':'en',
    'lr':'lang_en',
    'as_sdt':0,
    'num': 20,
    'start': 0
    }

for page in range(num_pages):
    print(f"retriving results from page [{page}]!",'\r')
    params['start'] += 20
    search = gs.GoogleScholarSearch(params)
    results = search.get_dict()


    # Extract the articles from the current page
    articles = results.get("organic_results", [])
    all_results.extend(articles)

    # Stop if there are no more results
    if not articles:
        print(f"No articles found!",'\r')
        break
    
    time.sleep(1.5)
```
